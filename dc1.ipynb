```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'], test_size=0.2)
```

# Lasso Testing

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

#lasso = Lasso(reg.best_params_)
params = {'alpha' : [1, 10, 15, 50, 100, 1000]}
lasso = Lasso()
lasso2 = GridSearchCV(lasso, params, scoring='r2', cv=5)

lasso2.fit(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'])
```

```python
lasso2.score(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'])
```

```python
lasso2.best_score_
```

# Testing best alpha value

```python
lasso2.best_params_
```

```python
lasso = Lasso(alpha=1)

lasso.fit(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'])
lasso.score(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'])
```

```python
r2_score(test_y, lasso.predict(test_x))
```

# Ridge Testing

```python
from sklearn.linear_model import Ridge

ridge = Ridge(normalize=True)
params = {'alpha' : [1, 10, 15, 50, 100, 1000]}
ridge2 = GridSearchCV(ridge, params, scoring='r2', cv=5)
```

```python
ridge2.fit(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'])
```

```python
ridge2.score(df.drop(columns=['Immigration to the United States','Region','Country']), df['Immigration to the United States'])
```

# Testing best alpha value

```python
ridge2.best_params_
```

```python
ridge2.best_score_
```

# The main difference between ridge and lasso is that lasso tends to make coefficients to absolute zero while ridge never sets the value of the coefficient to absolute zero

```python
r2_score(test_y, ridge2.predict(test_x))
```
